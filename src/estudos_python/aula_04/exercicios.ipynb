17. Crie uma função que receba um número como argumento e retorne True se o número for primo e False caso contrário.

````python
import math

def eh_primo(n) -> bool:
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    
    limite = int(math.sqrt(n)) + 1
    for i in range(3, limite, 2):
        if n % i == 0:
            return False
            
    return True

# Teste
numero = 29
print(f"O número {numero} é primo? {eh_primo(numero)}")
````

20. Escreva uma função que receba um dicionário e retorne uma lista de chaves ordenadas

````python
cadastro: dict = {"nome": "Alan", "idade": 32, "sexo": "Masculino", "natural": "Santos"}

lista_chaves: list = []

def chaves_ordenadas (dicionario: dict) -> list:
    for chave in dicionario.keys():
        lista_chaves.append(chave)
        lista_chaves.sort()
    return lista_chaves

resultado: list = chaves_ordenadas(cadastro)
print(resultado)
````
